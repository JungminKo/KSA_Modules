```python
# matplot 작성하기
x = np.arange(1, 11)
y = (x + np.random.randn(10)) * 5

plt.plot(x, y, "s") # 산점도, marker 출력
plt.show() # 그래프 출력
```

>### 기본 그래프 구성요소
- 제목과 축 레이블
    - 플롯에 X, Y 축 레이블이나 제목을 추가
            title()  : 차트의 제목
            xlabel() : 차트의 x-축 제목
            ylabel() : 차트의 y-축 제목
    - 옵션
        - `label` : 제목 텍스트
        - `fontdict` :
                {'fontsize':rcParams['axes.titlesize'], 
                 'fontweight':rcParams['axes.titleweight'],
                 'verticalalignment':'baseline', 
                 'horizontalalignment':loc}
        - `loc` : title 옵션으로 수평 정렬 (기본값 : "center")
                'center', 'left', 'right'
        - `labelpad` : xlabel과 ylabel 옵션으로 축과 축 제목 사이의 간격 (기본값 : None)

```python
plt.plot(x, y, "p")
plt.title('산점도 제목', loc='left',
         fontdict={'fontsize':25, 'fontweight':'bold'})
plt.xlabel('x-axis', labelpad=15)
plt.ylabel('y-axis', fontdict={'fontsize':10, 'fontweight':'bold'})
plt.show()
```

```python
# color='w' kargs를 넣으면 폰트 색깔을 바꿀 수 있다.
plt.plot(x, y, "p")
plt.title('산점도 제목', loc='right', color='r',
         fontdict={'fontsize':25, 'fontweight':'bold'})
plt.xlabel('x-axis', labelpad=15, color='r', 
           fontsize=15, fontweight='bold')
plt.ylabel('y-axis', color='r',
           fontdict={'fontsize':15, 'fontweight':'bold'})
plt.xticks(color='r', fontsize=15)
plt.yticks(color='r', fontsize=15)
plt.show()
```

```python
x = np.arange(1, 11)
y1 = (x + np.random.randn(10)) * 5
y2 = (x + np.random.randn(10)) * 5 + 5

plt.plot(x, y1, 'p')
plt.plot(x, y2, 'rp')
plt.title('x-y1, x-y2 산점도', fontsize=18)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis', fontsize=15)
plt.legend(['남자', '여자'], loc='lower right')
plt.axis(xmin=0, xmax=15, ymin=0, ymax=100) # xlim, ylim에서 설정 가능
plt.minorticks_on()
plt.grid(True, which='major', axis='y')
plt.show()
```

```python
x = np.arange(10)
y1 = 1.5 + 2.7 * x + np.random.randn(10)*5
y2 = 5.5 + 2.7 * x + np.random.randn(10)*5

plt.plot(x,y1,'gp--')
plt.plot(x,y2,'rp-')
plt.title("산점도", fontdict = {'fontsize':15, 'fontweight':"bold"}, loc="left")
plt.xlabel("x-축 이름표",labelpad =15)
plt.ylabel("y-축 이름표")
plt.legend(["그룹1","그룹2"], loc = "lower right")
plt.axis(xmin=0, xmax=12, ymin = -2, ymax = 50)
plt.minorticks_on()
plt.grid(b=True, which='minor', axis='y') # which='minor' 옵션 차이
plt.show()
```

### 원 차트 (Pie Chart)
>- 범주(category)로 구분되는 데이터를 이용하여 범주별로 원조각을 작성
        pie() : 원 그래프 작성
- 옵션
    - `x` : 원조각의 값
    - `explode` : 특정 조각 빼기 (기본값 : None)
    - `labels` : 각 조각별 이름 (기본값 : None)
    - `colors` : 각 조각의 색 (기본값 : None)
        - 색깔의 약자나 이름
    - `autopct` : 퍼센트 표기 (기본값 : None)
        - "%format%pct" 형식으로 입력
    - `shadow` : 그림자 효과로 입체값 주기 (기본값 : None)

```python
print(AreaTbl)

plt.pie(AreaTbl, explode=(0, .2, 0, 0, 0), labels=AreaTbl.index,
        colors=['r','y','g','c','m'], autopct="%5.1f%%", shadow=True)
plt.show()
```


>### 상자 그래프 (Boxplot)
- 상자그림(상자도표) 중심과 퍼짐의 요약을 얻을 수 있음 (다섯 수치 요약)
- 자료의 중심점 위치 확인 및 분포 대칭성 판단
- 자료의 중심위치 확인

    - `x` : 상자 그래프를 작성할 데이터

    - `whis` : 안쪽 울타리 값 (기본값 : 1.5)
    - `labels` : 항목의 이름을 리스트로 입력 (기본값 : None)

>### 히스토그램 (Histogram)

## Plotly
>### plotly
- 퀘벡주 몬트리올에 본사가 있는 온라인 분석 및 데이터 시각화 도구
- plotly는 python과 Django 프레임워크를 사용하여 제작되었으며 프론트 엔드는 JavaScript와 시각화 라이브러리 D3.js, HTML 및 CSS를 사용
- 파일은 Amazon S3에서 호스팅
- 온라인 게시를 위해서는 다음문서를 참고
    - http://plot.ly/python/getting-started/#initialization-for-online-plotting
- plotly offline을 사용하면 그래프를 오프라인으로 생성하고 로컬로 저장(HTML)할 수 있으며 이렇게 하면 온라인에서 필요한 인증작업이 필요 없음
        import plotly # 임포트
        plotly.offline.plot # 이 명령어로 실행시키는게 핵심     
- 공부 자료
    - https://plot.ly/python/ipython-notebook-tutorial/

```python
import plotly.graph_objs as go
import plotly as ply

df = pd.read_csv('./data/Python_Data/ch02_data/APTData.csv',
                header=0, skiprows=7, index_col=None, encoding='utf-8')
df.head()
```

```python
# 원그래프
data = [go.Pie(text=AreaTbl.index, values=AreaTbl.values, opacity=.7)]

layout = go.Layout(title='원그래프') # 그래프 layout
figure = go.Figure(data=data, layout=layout) # 그래프 작성
ply.offline.plot(figure, filename=\
                "./data/Python_Data/ch02_data/piechart.html")
```

```python
# 막대그래프
data = [go.Bar(x=AreaTbl.index, y=AreaTbl.values)]

layout = go.Layout(title='막대그래프') # 그래프 layout
figure = go.Figure(data=data, layout=layout) # 그래프 작성
ply.offline.plot(figure, filename=\
                "./data/Python_Data/ch02_data/barchart.html")
```

```python
# 상자그래프
data = [go.Box(x=df.Area, y=df.Price, boxmean=True, fillcolor='yellow',
               jitter=True, notched=True)]

layout = go.Layout(title='상자그래프') # 그래프 layout
figure = go.Figure(data=data, layout=layout) # 그래프 작성
ply.offline.plot(figure, filename=\
                "./data/Python_Data/ch02_data/boxplot.html")
```

```python
# 히스토그램
data = [go.Histogram(x=df.Price, nbinsx=15, opacity=.5, text="개")]

layout = go.Layout(title='히스토그램') # 그래프 layout
figure = go.Figure(data=data, layout=layout) # 그래프 작성
ply.offline.plot(figure, filename=\
                "./data/Python_Data/ch02_data/histogram.html")
```

```python
# 산점도
data = [go.Scatter(x=df.Price, y=df.FAR, mode='lines+markers')]

layout = go.Layout(title='산점도') # 그래프 layout
figure = go.Figure(data=data, layout=layout) # 그래프 작성
ply.offline.plot(figure, filename=\
                "./data/Python_Data/ch02_data/scatter.html")
```

## Bokeh
>### Bokeh
- 웹 브라우저에서 시각화에 효과적인 파이썬 인터랙티브 시각화 라이브러리

- IPython Notebook이랑 호환이 잘 된다.
- http://bokeh.pydata.org/en/latest/index.html

```python
from bokeh.plotting import figure, output_file, output_notebook, show
```

```python
# Output to static HTML file
# Output_file("lines.html")

# output to notebook
output_notebook()

# Create a new plot with a titie and axis labels
p = figure(title='막대그래프', x_axis_label='x', y_axis_label='y')

p.hbar(y=[1,2,3,4,5], height=.5, right=AreaTbl.values)

# show the results
show(p)
```
