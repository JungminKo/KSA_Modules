# 1-1. 추천 시스템 개요<br>
## 추천시스템?
- 주변의 정보를 Decision making의 수단으로 활용하는 것
- 게으른 사람들을 위한...?

## 등장 배경 (추천 시스템이 중요해진 이유)
- 정보 과부하 (information overload) 문제를 해결하기 위한 방법으로 개인화 서비스가 대두
- 인간은 처리할 수 있는 정보량 이상으로 정보를 접하게 되면 오히려 학습과 의사결정을 방해

## 개인화 서비스가 중요해졌다!
- 개인화 서비스 중 하나인 추천시스템은 각 개인의 선호르 분석하여 저갑한 컨텐츠를 제공
- 어지간한 기업들에 전부 장착이 되어있다고 생각하면 된다.
- 개별 고객 선호도를 파악하여 개인벼로 차별화된 서비스 또는 컨텐츠를 제공하는 시스템
    1. 최신 분석 기술의 적용
    2. 고객 입장
    3. 판매자 입장

# 1-2. 대표적 추천 기법<br>
## 대표적 추천 기법 개요
1. 내용 기반 필터링
    - 아이템 자체를 직접 분석하는 것
    - 아이템과 아이템, 아이템과 사용자 선호도간 유사성을 토대로 추천
    - 구매했는지 여부는 고려하지 않는다.
2. 협업 필터링
    - 대규모의 기존 사용자 구매정보를 분석
    - 비슷한 성향의 사용자들이 선호한 항목 또는 유사한 항목을 추천
3. 잠재 요인 모델
    - 통계 기반의 모델링 방법들을 많이 사용
4. 하이브리드
    - 내용 기반 + 협업 기반
    
## 추천 시스템 종류
- Memory-Based/Similarity-Based vs Model-Based

#### 내용 기반 필터링
- 아이템의 건텐츠를 직접 분석하여 추천하는 기법
    - 목표 고객이 구매한 아이테모가 아이템 또는 아이템과 사용자 선호도 간의 유사성을 토대로 추천
- 아이템 자체일 수도, 아이템이 가지고 있는 특징으로 분석할 수 있다.

#### 협업 필터링
- 학계 및 산업계에서 가장 성공적으로 알려진 기법
    - 아마존에서 적용, 널리 알려진 기법
        - 사실 예전얘기, 딥러닝 적용 논문은 많이 나오는데 실제 적용 사례는 많이 나오지 않음
    - 모교 고객의 구매 기록을 바탕으로 유사한 사용자 또는 아이템을 탐색하여 추천하는 기법
    - 유사도 측정 대상에 따라 사용자 기반(User-Based CF)과 아이템 기반(Item-Based CF)로 구분
    
#### 잠재 요인 모델
- 희박한 구매 기록을 기반으로 사용자와 아이템의 잠재 요인 탐색
    - Factoize, SVD를 사용, 사용자와 아이템을 나타내는 잠재 요인 모델 생성
    - Netflix Prize에서 가장 높은 성과를 보인 모델

#### 하이브리드 기법
- CF와 CB의 자엄을 결합하는 기법
    1. 가중평균합
    2. CF의 특성을 CB에 포함
    3. 통합 모델
    4. CB의 특성을 CF에 포함

## 추천 시스템 주요 이슈
1. 신규 고객 및 상품 (Cold Start)
2. 데이터의 희박성 (Sparsity)
3. 확장성 (Scalability)


# 1-3. 추천 시스템 평가<br>
## 추천 시스템의 성과 측정 방법
- 직접 실험의 어려움 : 머신러닝 기법의 평가와 유사하게 Train/Test Split 기법을 사용

## 사용자 선호 데이터의 구분
- 사용자의 선호를 직접적으로 알 수 있는 경우와 아닌 경우가 존재
    1. 명시적 선호(Explicit rating) : 평점 등 사용자의 선호가 수치로 존재
    2. 암시적 선호(Implicit rating) : 구매 기록 등 사용자의 선호를 간접적으로 측정할 수 있는 데이터

#### 명시적 선호 데이터의 특징
- 사용자의 선호를 추론하는 과정 불필요
- 명시적 선호 데이터를 남기는 고객이 극 소수
- 명시적 선호 데이터의 왜곡 가능성

#### 암시적 선호 데이터의 특징
- 명시적 선호 데이터 획득의 어려움으로 암시적 선호 데이터를 활용하려는 노력이 지속적으로 발전
- 구매 기록 등의 로그 데이터로부터 선호를 추론하는 모델이 필요
- Information Utilization Problem : 과연 데이터가 선호를 나타낼 수 있는가?

## 추천 시스템의 평가 기준
1. Accuracy : 정확한 추천을 제공
2. Diversity : 다양한 아이템 추천
3. Serendipity : 우연성
4. Speed : 소요된 시간

#### 명시적 선호 데이터의 정확도 측정 방법
- 예측값과 실제값의 오차를 측정
- RMSE (Root Mean Squared Error)
- MAE (Mean Absolulte Error)
- FCP (Fraction of Concordant Pairs) : 평점에 의한 순위를 기준으로 오차 측정

#### 구매 기록 데이터의 정확도 측정 방법
- precision
- recall
- F1
- accuracy
- specificity

# Design Thinking
- Definition
    - Stanford University, 앉아서 생각만 해서는 그 사람의 생각이 100% 나올 수 없다.
    - Creative
    - Emphaize > Define > Ideate > Prototype > Test
- Emphasize : 정보 수집이 목적
- Define : 정보 축소가 목적, 서로를 설득하는 과정
    - 어떤 방법론을 설정하는 것이 유용한지
- Ideate : 마인드맵, 브레인 스토밍, 바디 스토밍
- Prototype : 허접해야함. 왜냐, 최대한 빨리 만들어야 함
- Test
