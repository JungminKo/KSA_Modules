```python
import tensorflow as tf
tf.set_random_seed(777)

from keras.datasets import mnist
from keras.utils import to_categorical

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape((60000, 28, 28, 1))
X_train = X_train.astype('float32') / 255

X_test = X_test.reshape((10000, 28, 28, 1))
X_test = X_test.astype('float32') / 255

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, AvgPool2D
from keras.layers import LeakyReLU

model = Sequential()
# layer 1 Conv2D ---------------------------------------
model.add(Conv2D(32, (3, 3), padding='SAME',
                 input_shape=(28, 28, 1)))
model.add(LeakyReLU(alpha=.1))
model.add(MaxPooling2D(pool_size=(2, 2), padding='SAME',
                       strides=(2, 2)))
# layer 2 Conv2D ---------------------------------------
model.add(Conv2D(64, kernel_size=(5, 5), padding='SAME',
                 strides=(2, 2)))
model.add(LeakyReLU(alpha=.1))
model.add(MaxPooling2D(pool_size=(2, 2), padding='SAME',
                       strides=(2, 2)))
# layer 3 Conv2D ---------------------------------------
model.add(Conv2D(256, kernel_size=(3, 3), padding='SAME',
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), padding='SAME',
                       strides=(2, 2)))
# layer 4 Conv2D ---------------------------------------
model.add(Conv2D(128, kernel_size=(3, 3), padding='SAME',
                 activation='relu'))
model.add(AvgPool2D(pool_size=(2, 2), padding='SAME',
                    strides=(2, 2)))
# layer 5 Fully Connected ------------------------------
model.add(Flatten())
model.add(Dense(256))
# Output layer Conv2D ---------------------------------------
model.add(Dense(10, activation='softmax'))

model.summary()

import keras.backend as K

def myAccuracy(y_true, y_pred):
    diff = K.abs(y_true-y_pred) # absolute difference between correct and predicted values
    correct = K.less(diff,0.05) # tensor with 0 for false values and 1 for true values
    return K.mean(correct) # sum all 1's and divide by the total. 

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=[myAccuracy])
model.fit(X_train, y_train, epochs=20, batch_size=200)

test_loss, test_acc = model.evaluate(X_test, y_test)
test_acc # 이겼다 제성이형 ㅎㅎㅎㅎ
```
